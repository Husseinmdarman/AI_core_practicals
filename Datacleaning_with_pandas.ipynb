## Q6. Create a new column with the duration of each call in seconds

```python
car_insurance_dataframe[['CallStart', 'CallEnd']].head()
```

```python
car_insurance_dataframe['CallStart'] = pd.to_datetime(car_insurance_dataframe['CallStart'], format = '%H:%M:%S')
car_insurance_dataframe['CallEnd'] = pd.to_datetime(car_insurance_dataframe['CallEnd'], format = '%H:%M:%S')
```

```python
car_insurance_dataframe['Call_Duration'] = car_insurance_dataframe['Call_Duration'].apply(lambda x: x.total_seconds())
```

```python
car_insurance_dataframe
```

## Q9. What is the average duration of each call?

```python
car_insurance_dataframe['Call_Duration'].mean()
```

```python
print('the average duration of the call in seconds is {:.2f}'.format(car_insurance_dataframe['Call_Duration'].mean()))
```

## Q10. For those people contacted during the first half of the year (Jan-June). What is the most common way of comunication (telephone, cellular...)?

```python
car_insurance_dataframe['LastContactMonth'].astype('category')
```

```python
car_insurance_dataframe['LastContactMonth'].unique()
```

```python
months_to_check = ['jan', 'feb', 'mar','apr','may','jun']
```

```python
mask = car_insurance_dataframe['LastContactMonth'].apply(lambda x: x in months_to_check)
```

```python
most_common_communication_first_half_of_year = car_insurance_dataframe[mask]['Communication'].value_counts().idxmax()
```

```python
print(f'The most common communication form for the first half of the year is {most_common_communication_first_half_of_year}')
```
