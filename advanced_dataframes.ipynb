Question 4. Create a new column called "last_updated" where every value is equal to the current time (use the datetime module and format the date in ISO format).

```python
from datetime import datetime
```

```python
now = datetime.now()
current_time = now.strftime("%H:%M:%S")
salaries_dataframe['last_updated'] = current_time
```

Question 5. Create a new derived column called "time_ratio" in the DataFrame which is the proportion of overtime pay to base pay

The way i understood this question was that it wants to know the increase of percentage of overtime compared to base salary

```python
import math
```

```python
salaries_dataframe['time_ratio'] = (salaries_dataframe['OvertimePay'] / salaries_dataframe['BasePay']) * 100
```

```python
salaries_dataframe['time_ratio']= salaries_dataframe['time_ratio'].apply(lambda x: math.ceil(x))
```
